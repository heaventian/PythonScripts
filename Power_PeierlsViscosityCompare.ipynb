# Peierls plastic creep viscosity compared to dislocation viscosity
Parameters of Peierls plastic creep viscosity is taken from Katayama and Karato, 2008
Mantle viscosity is calculated by solving the strain_rate-stress equation based on the brent method
Note that Peierls stress is divided by a value of 1e21 to increase its viscosity.

```python
from scipy.optimize import brentq
import numpy as np
import matplotlib.pyplot as plt

T=np.linspace(300,1700,100)

E=530e3 # J/mol activation energy
V=13e-6 # m**3/mol activation volume
R=8.31 # J/K/mol
n=3.5
A=2.75e-16

SP=9.1e9 # Pa Peierls stress
AP = 6.3e-5/1e21
p=1.
q=2.

def ViscFn(A,E,P,V,n,R,T,SR):
    tmp1=(E+P*V)/(n*R*T)
    return A**(-1./n)*np.exp(tmp1)*SR**(1./n-1.)

def Fx_Peierls(x,SE,p,q,E,V,A,n,R,P,SR,T):
    x=min(x,SE)
    tmp1=(1-(x/SE)**p)**q
    tmp2=-(E+P*V)/(n*R*T)*tmp1
    y=SR-A*x**2*np.exp(tmp2)
    return y


plt.figure(figsize=(16,8))

P=1e9
SR=1e-12
eta1=ViscFn(A,E,P,V,n,R,T,SR)
S2=np.zeros_like(T)
eta2=np.zeros_like(T)
for i1 in range(len(S2)):
    S2[i1]=brentq(Fx_Peierls,0.0,1e16,maxiter=1000,args=(SP,p,q,E,V,AP,n,R,P,SR,T[i1]))
    eta2[i1]=S2[i1]/(2*SR)


plt.subplot(131)
plt.semilogy(T,eta1,'b-')
plt.plot(T,eta2,'b--')

SR=1e-15
eta1=ViscFn(A,E,P,V,n,R,T,SR)
S2=np.zeros_like(T)
eta2=np.zeros_like(T)
for i1 in range(len(S2)):
    S2[i1]=brentq(Fx_Peierls,0.0,1e16,maxiter=1000,args=(SP,p,q,E,V,AP,n,R,P,SR,T[i1]))
    eta2[i1]=S2[i1]/(2*SR)


plt.plot(T,eta1,'r-')
plt.plot(T,eta2,'r--')

SR=1e-18
eta1=ViscFn(A,E,P,V,n,R,T,SR)
S2=np.zeros_like(T)
eta2=np.zeros_like(T)
for i1 in range(len(S2)):
    S2[i1]=brentq(Fx_Peierls,0.0,1e16,maxiter=1000,args=(SP,p,q,E,V,AP,n,R,P,SR,T[i1]))
    eta2[i1]=S2[i1]/(2*SR)


plt.plot(T,eta1,'g-')
plt.plot(T,eta2,'g--')
plt.grid('on')
plt.axis([300,1700,1e17,1e30])
plt.title('Pressure = '+str(P/1e9)+' GPa')
plt.xlabel('Temperature (K)')
plt.ylabel('Viscosity (Pas)')

P=3e9
SR=1e-12
eta1=ViscFn(A,E,P,V,n,R,T,SR)
S2=np.zeros_like(T)
eta2=np.zeros_like(T)
for i1 in range(len(S2)):
    S2[i1]=brentq(Fx_Peierls,0.0,1e16,maxiter=1000,args=(SP,p,q,E,V,AP,n,R,P,SR,T[i1]))
    eta2[i1]=S2[i1]/(2*SR)


plt.subplot(132)
plt.semilogy(T,eta1,'b-')
plt.plot(T,eta2,'b--')

SR=1e-15
eta1=ViscFn(A,E,P,V,n,R,T,SR)
S2=np.zeros_like(T)
eta2=np.zeros_like(T)
for i1 in range(len(S2)):
    S2[i1]=brentq(Fx_Peierls,0.0,1e16,maxiter=1000,args=(SP,p,q,E,V,AP,n,R,P,SR,T[i1]))
    eta2[i1]=S2[i1]/(2*SR)


plt.plot(T,eta1,'r-')
plt.plot(T,eta2,'r--')

SR=1e-18
eta1=ViscFn(A,E,P,V,n,R,T,SR)
S2=np.zeros_like(T)
eta2=np.zeros_like(T)
for i1 in range(len(S2)):
    S2[i1]=brentq(Fx_Peierls,0.0,1e16,maxiter=1000,args=(SP,p,q,E,V,AP,n,R,P,SR,T[i1]))
    eta2[i1]=S2[i1]/(2*SR)


plt.plot(T,eta1,'g-')
plt.plot(T,eta2,'g--')
plt.grid('on')
plt.axis([300,1700,1e17,1e30])
plt.title('Pressure = '+str(P/1e9)+' GPa')
plt.xlabel('Temperature (K)')

P=1e10
SR=1e-12
eta1=ViscFn(A,E,P,V,n,R,T,SR)
S2=np.zeros_like(T)
eta2=np.zeros_like(T)
for i1 in range(len(S2)):
    S2[i1]=brentq(Fx_Peierls,0.0,1e16,maxiter=1000,args=(SP,p,q,E,V,AP,n,R,P,SR,T[i1]))
    eta2[i1]=S2[i1]/(2*SR)


plt.subplot(133)
plt.semilogy(T,eta1,'b-')
plt.plot(T,eta2,'b--')

SR=1e-15
eta1=ViscFn(A,E,P,V,n,R,T,SR)
S2=np.zeros_like(T)
eta2=np.zeros_like(T)
for i1 in range(len(S2)):
    S2[i1]=brentq(Fx_Peierls,0.0,1e16,maxiter=1000,args=(SP,p,q,E,V,AP,n,R,P,SR,T[i1]))
    eta2[i1]=S2[i1]/(2*SR)


plt.plot(T,eta1,'r-')
plt.plot(T,eta2,'r--')

SR=1e-18
eta1=ViscFn(A,E,P,V,n,R,T,SR)
S2=np.zeros_like(T)
eta2=np.zeros_like(T)
for i1 in range(len(S2)):
    S2[i1]=brentq(Fx_Peierls,0.0,1e16,maxiter=1000,args=(SP,p,q,E,V,AP,n,R,P,SR,T[i1]))
    eta2[i1]=S2[i1]/(2*SR)


plt.plot(T,eta1,'g-')
plt.plot(T,eta2,'g--')
plt.grid('on')
plt.axis([300,1700,1e17,1e30])
plt.title('Pressure = '+str(P/1e9)+' GPa')
plt.xlabel('Temperature (K)')

plt.legend(['SR=1E-12, Dislocation','SR=1E-12, Peierls','SR=1E-15, Dislocation','SR=1E-15, Peierls','SR=1E-18, Dislocation','SR=1E-18, Peierls'])
plt.show()
```

# Garel et al., 2014 way of approximate Peierls plastic viscosity
A large exponent (n=20) is used to approximate Peierls plastic viscosity.
Note that because viscosity can be explicitly expressed by strain rate, 
you do not need to solve an equation to get viscosity

```python
T=np.linspace(300,1700,100)

E=540e3 # J/mol activation energy
VP=12e-6 # m**3/mol activation volume
VE=10e-6 # m**3/mol activation volume
R=8.31 # J/K/mol
nP=3.5
nE=20
AP=5.0e-16
AE=1E-150


def ViscFn(A,E,P,V,n,R,T,SR):
    tmp1=(E+P*V)/(n*R*T)
    return A**(-1./n)*np.exp(tmp1)*SR**(1./n-1.)

plt.figure(figsize=(16,8))

P=1e9
SR=1e-12
eta1=ViscFn(AP,E,P,VP,nP,R,T,SR)
eta2=ViscFn(AE,E,P,VE,nE,R,T,SR)

plt.subplot(131)
plt.semilogy(T,eta1,'b-')
plt.plot(T,eta2,'b--')

SR=1e-15
eta1=ViscFn(AP,E,P,VP,nP,R,T,SR)
eta2=ViscFn(AE,E,P,VE,nE,R,T,SR)

plt.plot(T,eta1,'r-')
plt.plot(T,eta2,'r--')

SR=1e-18
eta1=ViscFn(AP,E,P,VP,nP,R,T,SR)
eta2=ViscFn(AE,E,P,VE,nE,R,T,SR)

plt.plot(T,eta1,'g-')
plt.plot(T,eta2,'g--')
plt.grid('on')
plt.axis([300,1700,1e18,1e30])
plt.title('Pressure = '+str(P/1e9)+' GPa')
plt.xlabel('Temperature (K)')
plt.ylabel('Viscosity (Pas)')

P=3e9
SR=1e-12
eta1=ViscFn(AP,E,P,VP,nP,R,T,SR)
eta2=ViscFn(AE,E,P,VE,nE,R,T,SR)

plt.subplot(132)
plt.semilogy(T,eta1,'b-')
plt.plot(T,eta2,'b--')

SR=1e-15
eta1=ViscFn(AP,E,P,VP,nP,R,T,SR)
eta2=ViscFn(AE,E,P,VE,nE,R,T,SR)

plt.plot(T,eta1,'r-')
plt.plot(T,eta2,'r--')

SR=1e-18
eta1=ViscFn(AP,E,P,VP,nP,R,T,SR)
eta2=ViscFn(AE,E,P,VE,nE,R,T,SR)

plt.plot(T,eta1,'g-')
plt.plot(T,eta2,'g--')
plt.grid('on')
plt.axis([300,1700,1e18,1e30])
plt.title('Pressure = '+str(P/1e9)+' GPa')
plt.xlabel('Temperature (K)')


P=1e10
SR=1e-12
eta1=ViscFn(AP,E,P,VP,nP,R,T,SR)
eta2=ViscFn(AE,E,P,VE,nE,R,T,SR)

plt.subplot(133)
plt.semilogy(T,eta1,'b-')
plt.plot(T,eta2,'b--')

SR=1e-15
eta1=ViscFn(AP,E,P,VP,nP,R,T,SR)
eta2=ViscFn(AE,E,P,VE,nE,R,T,SR)

plt.plot(T,eta1,'r-')
plt.plot(T,eta2,'r--')

SR=1e-18
eta1=ViscFn(AP,E,P,VP,nP,R,T,SR)
eta2=ViscFn(AE,E,P,VE,nE,R,T,SR)

plt.plot(T,eta1,'g-')
plt.plot(T,eta2,'g--')
plt.grid('on')
plt.axis([300,1700,1e18,1e30])
plt.title('Pressure = '+str(P/1e9)+' GPa')
plt.xlabel('Temperature (K)')


plt.legend(['SR=1E-12, Dislocation','SR=1E-12, Peierls','SR=1E-15, Dislocation','SR=1E-15, Peierls','SR=1E-18, Dislocation','SR=1E-18, Peierls'])
plt.show()
```
